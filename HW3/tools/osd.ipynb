```python
import torch
import numpy as np
from torchsummary import summary

model_path = "C:/Users/OmidTavasolifar/Documents/YolosPretraind Models/YolosPretraind Models/yolo_nas_s.pt"
model = torch.load(model_path)

if isinstance(model, list) or isinstance(model, dict):
    if isinstance(model, list):
        model = model[0]
    elif isinstance(model, dict):
        model = model[list(model.keys())[0]]

model.eval()

input_size = (3, 224, 224)

try:
    dummy_input = torch.randn(1, *input_size)
    output = model(dummy_input)
    if isinstance(output, list):
        print("The model's forward function is returning a list. Adjusting to use the first element.")
        output = output[0]

    print(f"Output shape: {output.shape}")
except Exception as e:
    print(f"Error during model forward pass: {e}")

summary(model, input_size)
```

```python
import torch
from torchvision import transforms
from PIL import Image

# Define the transformations for the input data
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load an image
img = Image.open('C:/Users/OmidTavasolifar/Documents/sahi-main/GTAV_SmallObjectDetection_Frame_11.jpg')
img = transform(img)
img = img.unsqueeze(0)  # Add batch dimension

# Predict using the model
with torch.no_grad():
    outputs = model(img)
    _, predicted = torch.max(outputs, 1)

print(f'Predicted class: {predicted.item()}')
```
